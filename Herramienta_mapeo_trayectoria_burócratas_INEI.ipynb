# MAPEO DE FUNCIONARIOS CLAVE PARA INVESTIGACIÓN DE TESIS

El cuaderno explora la trayectoria de distintos funcionarios del INEI con el uso de BD de Transparencia Personal. Esta herramienta contribuye en la identificación de personal clave según cargos y sus antecedentes en la institución.

```python
import pandas as pd
import gdown

# URLs de Google Drive (convertidas a formato de descarga directa)
url_feb_2025 = "https://drive.google.com/uc?export=download&id=1mWcVbEHXXaiioURfplVDKKBV-Q2ewLrs"
url_may_2024 = "https://drive.google.com/uc?export=download&id=10yQWmS5TQyT1HWqSEwTUDWmpL-cLXjPV"

# Descargar los archivos
gdown.download(url_feb_2025, "2025_feb.xlsx", quiet=False)
gdown.download(url_may_2024, "2024_may.xlsx", quiet=False)

# Cargar los archivos en dataframes
df_feb_2025 = pd.read_excel("2025_feb.xlsx")
df_may_2024 = pd.read_excel("2024_may.xlsx")
```

```python
# Paso 3: Hacer el merge con la data de Febrero 2025 para encontrar los cargos actuales
df_seguimiento = df_may_filtrado.merge(df_feb_2025, on=columnas_clave, how="left")

# Paso 4: Seleccionar solo las columnas relevantes
df_seguimiento = df_seguimiento[
    ["VC_PERSONAL_PATERNO", "VC_PERSONAL_MATERNO", "VC_PERSONAL_NOMBRES",
     "VC_PERSONAL_CARGO_x", "VC_PERSONAL_DEPENDENCIA_x",  # Cargo y dependencia en 2024
     "VC_PERSONAL_CARGO_y", "VC_PERSONAL_DEPENDENCIA_y"]  # Cargo y dependencia en 2025
]

# Renombrar columnas para mayor claridad
df_seguimiento.columns = [
    "Apellido Paterno", "Apellido Materno", "Nombres",
    "Cargo Mayo 2024", "Dependencia Mayo 2024",
    "Cargo Febrero 2025", "Dependencia Febrero 2025"
]
```
